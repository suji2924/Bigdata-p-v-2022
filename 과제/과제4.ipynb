    - 사고유형_소범주

[데이터 확인 및 처리 / 대범주 변수 생성]

```python
df['사고유형'].value_counts()
```

```python
df['노면상태'].value_counts()
```

```python
df['도로형태'].value_counts()
```

```python
df['사고유형'].str.split(' - ', expand=True )
```

```python
#사고유형 대범주 변수 생성
df['사고유형_대범주'] = df['사고유형'].str.split(' - ').str[0]
df.head()
```

```python
#노면상태 대범주 변수 생성
df['노면상태_대범주'] = df['노면상태'].str.split(' - ').str[0]
df.head()
```

```python
#도로형태 대범주 변수 생성
df['도로형태_대범주'] = df['도로형태'].str.split(' - ').str[0]
df.head()
```

[데이터 처리 후 변수 이름 수정]

```python
df = df.rename(columns= {'사고유형':'사고유형_소범주', '노면상태':'노면상태_소범주', '도로형태':'도로형태_소범주'})
df.head()
```
