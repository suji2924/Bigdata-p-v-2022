## <span style="color:#004f9f"> 2. Filtering </span>

### <span style="color:#004f9f"> 2.1. Boolean Series </span>

**[비교 및 논리연산자]**

```python
# 2020년도 여부
df['Y2020'] = df['사고년도']==2020
```

```python
# 사고요일이 주말?
df['사고요일']>=5
```

```python
(df['사고요일']>=5).sum()
```

```python
(df['사고요일']>=5).mean()
```

```python
# 법규위반이 중앙선침범?
df['법규위반'] == '중앙선침범'
```

```python
# 주말이고 기상상태가 맑음?
c1 = df['사고요일']>=5
c2 = df['기상상태'] == '맑음'
(c1 & c2).mean()
```

```python
# 중상자수가 5명이상이거나 사망자수가 1명이상?
c1 = df['중상자수']>=5
c2 = df['사망자수']>=1
(c1 | c2).sum()
```

```python
df['사고년도'].value_counts()
```

```python
# 2019년도의 강남구?
c1 = ~df['Y2020']
c2 = df['시군구_대범주'] == '강남구'
(c1 & c2).sum()
```

```python
# 강남구와 성북구의 2019년도?
c1 = df['시군구_대범주'] == '강남구'
c2 = df['시군구_대범주'] == '성북구'
((c1 | c2) & (~df['Y2020'])).sum()
```

```python
#이게 더 좋은 풀이 
my_region = ['강남구','성북구']
c1 = (df['시군구_대범주'].isin(my_region))
(c1 &(~df['Y2020'])).sum()
```

```python
# 가해운전자연령이 20대?
c1 = df['가해운전자연령']>=20
c2 = df['가해운전자연령']<=29
(c1 & c2).sum()
```

```python
df['가해운전자연령'].between(20,29).sum()
```

### <span style="color:#004f9f"> 2.2. 필터링 사례</span>

```python
# 강남구와 성북구의 2019년도 사고만 필터링
my_region = ['강남구','성북구']
c = (df['시군구_대범주'].isin(my_region))
c = (c & (~df['Y2020']))
#df[c]
df.loc[c]
```

```python
# 강남구와 성북구의 2019년도 사고의 부상신고자수
my_region = ['강남구','성북구']
my_col = ['부상신고자수']
c = (df['시군구_대범주'].isin(my_region))
c = (c & (~df['Y2020']))
df.loc[c, my_col]
```

```python
# 가해운전자연령이 20대인 사고의 수치 요약
df[df['가해운전자연령'].between(20,29)].describe().T
#qur = "20<=가해운전자연령<=29"
#df.query(qur).describe().T
```

```python
# 가해운전자연령이 80대이상 사고의 수치 요약
df[df['가해운전자연령']>=80].describe().T
#qur = "80<=가해운전자연령"
#df.query(qur).describe().T
```

### <span style="color:#004f9f"> 3.1. 단순집계</span>

### <span style="color:#004f9f"> 3.2. 단일열 기준 집계</span>

**[요일별 중상자수의 합계]**

```python
# agg 메서드 사용
df.groupby(['사고요일'])[['중상자수']].agg('sum')
```

```python
# 직접 집계함수의 메서드 사용
df.groupby(['사고요일'])[['중상자수']].sum()
```

```python
# agg와 딕셔너리 사용
df.groupby(['사고요일']).agg({'중상자수':'sum'})
```

**[요일별 부상신고자수, 중상자수의 합계]**

```python
df.groupby(['사고요일'])[['부상신고자수','중상자수']].sum()
```

**[기상상태별 부상신고자수의 합계, 사고건수]**

```python
df.groupby(['기상상태']).agg({'부상신고자수':['sum','count']})
```

**[사고요일별 사고건수, 가해운전자연령의 평균, 중상자수의 합계]**

```python
my_dict = {'가해운전자연령':['count','mean'],
           '중상자수': 'sum'}
my_agg = df.groupby(['사고요일']).agg(my_dict)
my_agg.columns= ['사고건수','가해운전자연령평균','중상자수합계']
my_agg['사고건당중상자수'] = my_agg['중상자수합계'] / my_agg['사고건수']
my_agg
```

### <span style="color:#004f9f"> 3.3. 복수열 기준 집계</span>

**[요일별 사고유형별 중상자수의 평균]**

```python
df.groupby(['사고요일','사고유형_대범주'])[['중상자수']].mean()
```

```python
df.groupby(['사고요일','사고유형_대범주'])[['중상자수']].mean().reset_index()
```

### <span style="color:#004f9f"> 3.4. Unstack과 Pivot Table</span>

**[unstack]**

```python
my_agg = df.groupby(['사고요일','사고유형_대범주'])[['중상자수']].mean()
my_agg
```

```python
my_agg.unstack()
```

**[pivot_table]**

```python
pd.pivot_table(df, index='사고요일', columns='사고유형_대범주', values='중상자수', aggfunc='mean', fill_value=0)
```

```python
pd.pivot_table(df, index=['사고요일','사고유형_대범주'], columns='가해운전자성별', values='중상자수', aggfunc='mean', fill_value=0)
```
